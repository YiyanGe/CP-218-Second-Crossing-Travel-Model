```python
#(1) single-occupancy passenger 
#; vehicles, or drive alone, "DA"; (2) two-occupant passenger vehicles, or shared ride two, "S2"; (3) three-or-more-
#; occupant passenger vehicles, or shared ride three, "S3"; (4) very small commercial trucks, which are assumed to 
#; be two-axle vehicles, "VSM"; (5) small commercial trucks, which are assumed to be two-axle vehicles, "SML"; (6)
#; medium commercial trucks, which are assumed to be three-axle vehicles, "MED"; and, (7) combination trucks, which
#; are charged the average of the five- and six-axle fee, "LRG". 
#http://analytics.mtc.ca.gov/foswiki/Main/LoadedHighway
modes = ['DA','S2','S3','SM','HV']
#modes = ['DA']
output = pd.DataFrame(index=['DA','S2','S3','SM','HV'],columns=['EA','AM','MD','PM','EV'])
timeperiod = ['EA','AM','MD','PM','EV']
#timeperiod=['AM']

for tp in timeperiod:
    load = pd.DataFrame.from_csv('test_data/avgload{}_link.csv'.format(tp))
    load['A'] = load.index
    load.reset_index(drop=True,inplace=True)
    load_baybridge = load[(load['TOLLCLASS'] == 5)]
    for mode in modes:
        load_baybridge['VOL_TTL_{}'.format(mode)] = load_baybridge[['VOL_{}'.format(mode),'VOL_{}T'.format(mode)]].apply(lambda x: x[0]+x[1], axis=1)
        if mode == 'SM':
            revenue=load_baybridge[['TOLL{0}_SML'.format(tp,mode),'VOL_TTL_{}'.format(mode)]].apply(lambda x: x[0]*x[1],axis=1).sum()
        elif mode == 'HV':
            revenue=load_baybridge[['TOLL{0}_LRG'.format(tp,mode),'VOL_TTL_{}'.format(mode)]].apply(lambda x: x[0]*x[1],axis=1).sum()
        else:
            revenue=load_baybridge[['TOLL{0}_{1}'.format(tp,mode),'VOL_TTL_{}'.format(mode)]].apply(lambda x: x[0]*x[1],axis=1).sum()

        output.set_value(mode, tp, revenue)
```
